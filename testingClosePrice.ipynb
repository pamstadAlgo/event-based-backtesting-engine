```python
import pandas_datareader.data as web
from datetime import date

df = web.DataReader("MULT.DE", "stooq").sort_index()
target = date(2025, 7, 31)

try:
    close = df.loc[target.isoformat(), "Close"]
    print(f"AAPL close on {target}: {close}")
except KeyError:
    print("No price for that date")
```
