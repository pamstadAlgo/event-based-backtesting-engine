```python
import pandas as pd

def download_stooq(symbol: str, interval: str = "d") -> pd.DataFrame:
    symbol = symbol.lower()
    url = f"https://stooq.com/q/d/l/?s={symbol}&i={interval}"
    df = pd.read_csv(url)  # Stooq returns a CSV


    print(df)
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date").set_index("Date")
    return df

df = download_stooq("XPEL.US", interval="d")
print(df.head())
print(df.tail())
```

```python
import pandas as pd

path = 'stooq_daily_data/us/nasdaq stocks/3/wldn.us.txt'

df = pd.read_csv(path)
df.head()
```

```python
date_col_name = '<DATE>'

df[date_col_name] = pd.to_datetime(
    df[date_col_name].astype(str),
    format="%Y%m%d",
    errors="raise",
).dt.date

df = df.sort_values(date_col_name).set_index(date_col_name)

df.head()
```
